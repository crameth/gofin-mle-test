```python
import pandas as pd
```

```python
df = pd.read_csv('../data/cs_assignment')
```

```python
df.info()
```

```python
df.describe()
```

```python
df['default_flag'].unique()
```

```python
# drop unwanted columns
features = [
    'outstanding', 
    'credit_limit', 
    'total_cash_usage', 
    'total_retail_usage', 
    'remaining_bill', 
    'payment_ratio', 
    'overlimit_percentage',
    'payment_ratio_3month',
    'payment_ratio_6month',
    'delinquency_score',
    'total_usage',
    'remaining_bill_per_number_of_cards',
    'remaining_bill_per_limit',
    'total_usage_per_limit',
    'total_3mo_usage_per_limit',
    'total_6mo_usage_per_limit',
    'utilization_3month',
    'utilization_6month',
]

features = df[features]

target = df['default_flag']
```

```python
import xgboost as xgb
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

D_train = xgb.DMatrix(X_train, label=y_train)
D_test = xgb.DMatrix(X_test, label=y_test)
```

```python
param = {
    'eta': 0.3, 
    'max_depth': 5,  
    'objective': 'binary:logistic',
} 

steps = 20  # The number of training iterations

model = xgb.train(param, D_train, steps)
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, accuracy_score

preds = model.predict(D_test)
best_preds = np.asarray([np.argmax(line) for line in preds])

print("Precision = {}".format(precision_score(y_test, best_preds, average='macro')))
print("Recall = {}".format(recall_score(y_test, best_preds, average='macro')))
print("Accuracy = {}".format(accuracy_score(y_test, best_preds)))
```

```python
model.save_model('../model/xgb.json')
```
